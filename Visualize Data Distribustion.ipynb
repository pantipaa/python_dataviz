```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

### Scatter plot

```python
irisdata = pd.read_csv('C:\\Users\\admin\\Downloads\\iris.data',header=None)  #ไฟล์ที่ไม่มีหัวตาราง
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()   
setosa_np
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])  #ใช้เฉพาะx,y
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])  #เพิ่มขนาด
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c = 'g')  #การกำหนดสี
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap='YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=np.array([[1,0.5,0.8]]))
```

## visualize 2 data

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c=setosa_np[:,3],cmap='BuPu',marker='p')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2],c=versicolor_np[:,3],cmap='BuPu',marker='*')
```

## visualize 3 data

```python
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0], setosa_np[:,1], s = 20*setosa_np[:,2], c=setosa_np[:,3],cmap = 'YlOrBr',marker='p')
plt.scatter(versicolor_np[:,0], versicolor_np[:,1], s = 20*versicolor_np[:,2], c=versicolor_np[:,3],cmap = 'YlOrBr',marker='*')
plt.scatter(virginica_np[:,0], virginica_np[:,1], s = 20*virginica_np[:,2], c=virginica_np[:,3],cmap = 'YlOrBr',marker='X')
plt.show()
```
