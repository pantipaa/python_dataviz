```python
import pandas as pd
```

```python
irisdata = pd.read_csv('C:\\Users\\admin\\Downloads\\iris.data',header=None)  #ไฟล์ที่ไม่มีหัวตาราง
irisdata
```

```python
import numpy as np
```

```python
data_np = irisdata.iloc[:,:-1].to_numpy()    #เอาตัวเลขตั้งแต่ตังแรกถึงก่อนัวสุดท้าย และตัดเลเวล ตัดชื่อออก
data_np
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()   
setosa_np
```

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()   
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

## visualize array

```python
from matplotlib import pyplot as plt   #สร้างกราฟ วาดรูปต่างๆ
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = [20,20]  #คำสั่งเ่มขนาด
plt.imshow(data_np, interpolation='nearest',cmap=plt.get_cmap('Greys'))  #เปลี่ยนสี
plt.show()    #สว่างค่ามากเข้มค่าน้อย
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12]  
plt.subplot(1,3,1)
plt.imshow(setosa_np,interpolation='nearest')
plt.subplot(1,3,2)
plt.imshow(versicolor_np,interpolation='nearest')
plt.subplot(1,3,3)
plt.imshow(virginica_np,interpolation='nearest')
plt.show()
```
